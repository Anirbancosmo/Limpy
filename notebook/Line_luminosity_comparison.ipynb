```python
import numpy as np
import matplotlib.pyplot as plt
from plotsettings import *
#import limpy
import limpy.lline as ll
#from plotsettings import *
import limpy.theory as lt
import limpy.utils as lu
```

```python
Mh1 = [1e11]
Mh2 = [1e12]
Mh3 = [1e13]
line_name='CO10'
z=np.linspace(0.1,3)
lcos1 = ll.mhalo_to_lline(Mh1, z, line_name= line_name, use_scatter=False)
lcos2 = ll.mhalo_to_lline(Mh2, z, line_name= line_name, use_scatter=False)
lcos3 = ll.mhalo_to_lline(Mh3, z, line_name= line_name, use_scatter=False)
```

```python
lcot1=lt.mhalo_to_lline(Mh1,z, line_name= line_name)
lcot2=lt.mhalo_to_lline(Mh2,z, line_name= line_name)
lcot3=lt.mhalo_to_lline(Mh3,z, line_name= line_name)
```

```python
color=['r', 'g', 'b']
lw=2.0 
plt.plot(z, lcos1, lw=lw,  color=color[0])
plt.plot(z, lcos2, lw=lw,  color=color[1])
plt.plot(z, lcos3, lw=lw,  color=color[2])

plt.plot(z, lcot1, '--', lw=lw,  color=color[0])
plt.plot(z, lcot2,'--' ,lw=lw,  color=color[1])
plt.plot(z, lcot3, '--',lw=lw,  color=color[2])
plt.yscale("log")
plt.ylim(1e2,)

plt.ylabel(r"$L_{CO10}$")
plt.xlabel(r"$z$")
```
