```python
import numpy as np
import matplotlib.pyplot as plt
import limpy.lline as ll
```

```python
boxsize=500  # Size of the box in comoving Mpc (cMpc)
ngrid= 2048 # number of grids along the box axis
nproj=1    # no of cells to project
halo_redshift=7.6 # redshift of halos, (Note: at this point it is not used in the function)

#It can read halos and density only from the redshift information but now, lets just do it for one redshift 
# by manually providng the files

#location of density file
root='/Volumes/Maxtor/density_halos/'
#root1="/Users/anirbanroy/Documents/21cmFAST/Output_files/Halo_lists/"

#dens_file=root+'updated_smoothed_deltax_z00%1.2f_%d_%dMpc' %(halo_redshift, ngrid,boxsize)

#location of halo catalogue file

halocat_file=root+'halos_z%1.2f_%d_%dMpc.npz' %(halo_redshift, ngrid,boxsize)
```

```python
print(halocat_file)
```

```python
# plot CII luminosity using beam convolution. 
# Use_scatter= True

beam_fwhm=2
beam_unit='arcmin'
ll.plot_beam(beam_fwhm,beam_unit, boxsize, ngrid, nproj, halocat_file, halo_redshift, 
          halo_cutoff_mass=1e11, use_scatter=True, unit='degree', halocat_file_type='npz', add_noise=False, random_noise_parcentage=None)

plt.savefig("CII_z%1.2f_L80N512_f200_mmin1e11.pdf" %(halo_redshift))
```
